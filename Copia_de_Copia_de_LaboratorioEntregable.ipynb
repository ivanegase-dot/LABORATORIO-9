a. El sistema debe permitir ingresar las notas de los estudiantes mediante un
menú interactivo. Utiliza un while loop para mostrar el menú hasta que el
usuario decida salir.
b. Utiliza una lista vacía para almacenar las notas y los nombres de los estudiantes.
Laboratorio #2
c. Usa try-except para manejar errores en las entradas (por ejemplo, si el usuario
ingresa una letra en lugar de un número).
d. Los estudiantes deben ser identificados por su nombre, y las notas deben ser
almacenadas en la lista como tuplas o diccionarios

```python
students_grades = []
MAX_GRADE_ATTEMPTS = 3 # Define el número máximo de intentos para ingresar una nota válida al modificar

while True:
    print("\n--- Menú de Notas de Estudiantes ---")
    print("1. Ingresar nota de un estudiante")
    print("2. Mostrar todas las notas")
    print("3. Modificar nota de un estudiante") # Nueva opción
    print("4. Salir") # Opción Salir reubicada

    choice = input("Ingrese su opción: ")

    if choice == '1':
        student_name = input("Ingrese el nombre del estudiante: ")
        while True:
            try:
                grade_input = input(f"Ingrese la nota de {student_name}: ")
                student_grade = float(grade_input)
                if 0 <= student_grade <= 100:
                    break
                else:
                    print("La nota debe estar entre 0 y 100.")
            except ValueError:
                print("Entrada inválida. Por favor, ingrese un número para la nota.")

        # Almacenar como diccionario
        students_grades.append({'name': student_name, 'grade': student_grade})
        print(f"Nota de {student_name} ({student_grade}) guardada exitosamente.")

    elif choice == '2':
        if not students_grades:
            print("No hay notas ingresadas aún.")
        else:
            print("\n--- Notas de Estudiantes ---")
            total_grades = 0
            num_students = 0
            for student_data in students_grades:
                print(f"Nombre: {student_data['name']}, Nota: {student_data['grade']}")
                total_grades += student_data['grade']
                num_students += 1

            try:
                average_grade = total_grades / num_students
                print(f"\nEl promedio de las notas es: {average_grade:.2f}")
            except ZeroDivisionError:
                # This case should ideally be caught by 'if not students_grades:' but included for robustness.
                print("No se puede calcular el promedio porque no hay notas de estudiantes para promediar.")

    elif choice == '3': # Implementación de la nueva funcionalidad 'Modificar nota'
        print("\n--- Modificar Nota de Estudiante ---")
        student_name_to_modify = input("Ingrese el nombre del estudiante cuya nota desea modificar: ")

        student_found = False
        for i, student_data in enumerate(students_grades):
            if student_data['name'].lower() == student_name_to_modify.lower():
                student_found = True
                print(f"Estudiante '{student_name_to_modify}' encontrado. Nota actual: {student_data['grade']}")

                attempts_grade_input = 0
                while attempts_grade_input < MAX_GRADE_ATTEMPTS:
                    try:
                        new_grade_input = input(f"Ingrese la nueva nota para {student_name_to_modify} (Intentos restantes: {MAX_GRADE_ATTEMPTS - attempts_grade_input}): ")
                        new_grade = float(new_grade_input)
                        if 0 <= new_grade <= 100:
                            students_grades[i]['grade'] = new_grade
                            print(f"Nota de {student_name_to_modify} actualizada a {new_grade} exitosamente.")
                            break
                        else:
                            print("La nota debe estar entre 0 y 100.")
                    except ValueError:
                        print("Entrada inválida. Por favor, ingrese un número para la nota.")
                    attempts_grade_input += 1

                if attempts_grade_input == MAX_GRADE_ATTEMPTS:
                    print(f"Se excedió el número máximo de intentos para ingresar una nota válida para {student_name_to_modify}.")

                break

        if not student_found:
            print(f"Error: Estudiante '{student_name_to_modify}' no encontrado en la lista.")

    elif choice == '4': # Opción Salir
        print("Saliendo del sistema. ¡Hasta luego!")
        break

    else:
        print("Opción no válida. Por favor, intente de nuevo.")
```
