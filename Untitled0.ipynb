```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de ventas mensuales (en miles)
data = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto",
            "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90]
}

# Crear un DataFrame
df_ventas = pd.DataFrame(data)
display(df_ventas)
```

### Operaciones con NumPy y Estadísticas Básicas

Ahora, utilizaremos NumPy para calcular algunas estadísticas clave de las ventas.

```python
# Convertir la columna 'Ventas' a un array de NumPy para operaciones
ventas_np = df_ventas['Ventas'].to_numpy()

# Calcular la venta total
venta_total = np.sum(ventas_np)

# Calcular la venta promedio
venta_promedio = np.mean(ventas_np)

# Encontrar la venta máxima
venta_maxima = np.max(ventas_np)

# Encontrar la venta mínima
venta_minima = np.min(ventas_np)

print(f"Venta Total Anual: {venta_total} mil")
print(f"Venta Promedio Mensual: {venta_promedio:.2f} mil")
print(f"Venta Máxima Mensual: {venta_maxima} mil")
print(f"Venta Mínima Mensual: {venta_minima} mil")
```

### Visualización de Resultados con Matplotlib

Finalmente, visualizaremos las ventas mensuales para entender su tendencia.

```python
# Crear un gráfico de línea de las ventas mensuales
plt.figure(figsize=(12, 6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas'], marker='o', linestyle='-')
plt.title('Ventas Mensuales (en miles)')
plt.xlabel('Mes')
plt.ylabel('Ventas (en miles)')
plt.grid(True)
plt.xticks(rotation=45) # Rotar etiquetas del eje x para mejor legibilidad
plt.tight_layout() # Ajustar el diseño para evitar que las etiquetas se corten
plt.show()

# Crear un gráfico de barras de las ventas mensuales
plt.figure(figsize=(12, 6))
plt.bar(df_ventas['Mes'], df_ventas['Ventas'], color='skyblue')
plt.title('Ventas Mensuales (en miles)')
plt.xlabel('Mes')
plt.ylabel('Ventas (en miles)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de calificaciones de estudiantes
data_calificaciones = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah",
                   "Ian", "Jack"],
    "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

# Crear un DataFrame
df_calificaciones = pd.DataFrame(data_calificaciones)
display(df_calificaciones)
```

### Operaciones con NumPy y Estadísticas Básicas de Calificaciones

Ahora, utilizaremos NumPy para calcular algunas estadísticas clave de las calificaciones por asignatura y en general.

```python
# Calcular estadísticas para cada asignatura
asignaturas = ['Matemáticas', 'Ciencias', 'Historia']

print("Estadísticas por Asignatura:")
for asignatura in asignaturas:
    notas_np = df_calificaciones[asignatura].to_numpy()
    promedio_asignatura = np.mean(notas_np)
    maxima_asignatura = np.max(notas_np)
    minima_asignatura = np.min(notas_np)
    print(f"  {asignatura}:")
    print(f"    Promedio: {promedio_asignatura:.2f}")
    print(f"    Máxima: {maxima_asignatura}")
    print(f"    Mínima: {minima_asignatura}")

# Calcular el promedio general de todos los estudiantes
promedio_general = df_calificaciones[asignaturas].to_numpy().mean()
print(f"\nPromedio General de Calificaciones: {promedio_general:.2f}")
```

### Visualización de Resultados con Matplotlib

Finalmente, visualizaremos las calificaciones de los estudiantes y los promedios por asignatura.

```python
# Gráfico de barras de calificaciones por estudiante y asignatura
df_calificaciones.set_index('Estudiante').plot(kind='bar', figsize=(14, 7))
plt.title('Calificaciones de Estudiantes por Asignatura')
plt.xlabel('Estudiante')
plt.ylabel('Calificación')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Gráfico de barras de promedios por asignatura
promedio_por_asignatura = df_calificaciones[asignaturas].mean()

plt.figure(figsize=(10, 6))
promedio_por_asignatura.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Promedio de Calificaciones por Asignatura')
plt.xlabel('Asignatura')
plt.ylabel('Calificación Promedio')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
