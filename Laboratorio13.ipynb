Prompt 1: Análisis de Sentimiento con TextBlob

Instrucciones:
Utiliza la librería TextBlob para analizar el sentimiento de las siguientes frases. Debes interpretar el sentimiento de las frases como positivo, negativo o neutral en base a su polaridad.

Frase 1: "The performance optimization model significantly improved the athlete’s output during testing."
Frase 2: "The system design failed to meet the required endurance standards in the final evaluation"
Frase 3: "TThe results of the speed test were recorded for further analysis"

Objetivo:
Para cada frase, analiza la polaridad y asigna el sentimiento correspondiente:

Positivo si la polaridad es mayor que 0.
Negativo si la polaridad es menor que 0.
Neutral si la polaridad es 0.

```python
from textblob import TextBlob

# Frases a analizar
frases = [
    "The performance optimization model significantly improved the athlete’s output during testing.",
    "The system design failed to meet the required endurance standards in the final evaluation",
    "The results of the speed test were recorded for further analysis"
]

# Función para clasificar el sentimiento
def clasificar_sentimiento(polaridad):
    if polaridad > 0:
        return "Positivo"
    elif polaridad < 0:
        return "Negativo"
    else:
        return "Neutral"

# Análisis
for i, frase in enumerate(frases, 1):
    blob = TextBlob(frase)
    polaridad = blob.sentiment.polarity
    sentimiento = clasificar_sentimiento(polaridad)

    print(f"Frase {i}: {frase}")
    print(f"Polaridad: {polaridad}")
    print(f"Sentimiento: {sentimiento}")
    print("-" * 50)
```

Prompt 2: Entrenamiento de Modelo de Clasificación de Sentimientos con Naive Bayes

Instrucciones:
Entrena un modelo de clasificación de sentimientos utilizando el algoritmo Naive Bayes. Utiliza los siguientes ejemplos de frases etiquetadas y evalúa la precisión de tu modelo.

Datos de ejemplo:

Frase 1: "This app is very intuitive!"
Frase 2: "I didn't enjoy the customer service."
Frase 3: "Quite average experience."
Frase 4: "Fantastic presentation!"
Frase 5: "The food was awful."

Objetivo:

Entrena un modelo de Naive Bayes con estas frases etiquetadas.
Evalúa la precisión del modelo utilizando las frases proporcionadas.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Datos (frases + etiquetas)
frases = [
    "This app is very intuitive!",
    "I didn't enjoy the customer service.",
    "Quite average experience.",
    "Fantastic presentation!",
    "The food was awful."
]

# Etiquetas: positivo, negativo, neutral
etiquetas = ["positivo", "negativo", "neutral", "positivo", "negativo"]

# Vectorización (convertir texto a números)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(frases)

# Modelo Naive Bayes
modelo = MultinomialNB()
modelo.fit(X, etiquetas)

# Predicción sobre los mismos datos (evaluación simple)
predicciones = modelo.predict(X)

# Evaluar precisión
precision = accuracy_score(etiquetas, predicciones)

# Mostrar resultados
for i in range(len(frases)):
    print(f"Frase: {frases[i]}")
    print(f"Real: {etiquetas[i]} | Predicción: {predicciones[i]}")
    print("-" * 40)

print(f"Precisión del modelo: {precision}")
```
